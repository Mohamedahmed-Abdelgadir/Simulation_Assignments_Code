```python
import pandas as pd
```

```python
random_numbers_list = [53, 14, 75, 80, 50, 40, 49, 60, 78, 54, 17, 12, 15, 44, 55, 18, 9, 45, 24, 38, 97, 76, 89, 24, 66, 73, 39, 32, 4, 86]
```

```python
table = {
    "Day": [],
    "RN for news type": [],
    "News type": [],
    "RN for demand": [],
    "Demand": [],
    "Revenue": [],
    "Lost profit": [],
    "Cost": [],
    "Scrap": [],
    "Profit": []
}
```

```python
def determine_news_type(random_number):
  if (random_number in range(1, 36)):
    return "Good"

  elif (random_number in range(36, 81)):
    return "Fair"

  else:
    return "Poor"
```

```python
def determine_demand(random_number, news_type):
  if (news_type == "Good"):
    if(random_number in range(1, 4)):
      return 40
    elif (random_number in range(4, 9)):
      return 50
    elif (random_number in range(9, 24)):
      return 60
    elif (random_number in range(24, 44)):
      return 70
    elif (random_number in range(44, 79)):
      return 80
    elif (random_number in range(79, 94)):
      return 90
    else:
      return 100


  elif (news_type == "Fair"):
    if(random_number in range(1, 11)):
      return 40
    elif (random_number in range(10, 30)):
      return 50
    elif (random_number in range(30, 69)):
      return 60
    elif (random_number in range(69, 89)):
      return 70
    elif (random_number in range(89, 96)):
      return 80
    else:
      return 90


  else:
    if(random_number in range(1, 45)):
      return 40
    elif(random_number in range(45, 67)):
      return 50
    elif(random_number in range(67, 83)):
      return 60
    elif(random_number in range(83, 95)):
      return 70
    else:
      return 80
```

```python
def determine_lost_profit(demand):
  if (demand > 70):
    return (demand - 70) * 0.17
  else:
    return 0
```

```python
def determine_scrap(demand):
  if (demand < 70):
    return (70 - demand) * 0.07
  else:
    return 0
```

```python
def determine_revenue(demand):
  if (demand > 70):
    return 70 * 0.5
  else:
    return demand * 0.5
```

```python
for i in range(1, 70):
  table['Day'].append(i)
  table['RN for news type'].append(random_numbers_list.pop())
  table['News type'].append(determine_news_type(table['RN for news type'][i-1]))
  table['RN for demand'].append(random_numbers_list.pop())
  table['Demand'].append(determine_demand(table['RN for demand'][i-1], table['News type'][i-1]))
  table['Revenue'].append(determine_revenue(table['Demand'][i-1]))
  table['Lost profit'].append(determine_lost_profit(table['Demand'][i-1]))
  table['Cost'].append(70 * 0.33)
  table['Scrap'].append(determine_scrap(table['Demand'][i-1]))
  table['Profit'].append(table['Revenue'][i-1] - table['Cost'][i-1] - table['Lost profit'][i-1] + table['Scrap'][i-1])
```

```python
df = pd.DataFrame(table)
```
